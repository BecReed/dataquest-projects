# DataQuest Guided Project 4: College Major Data Visualization

This is the fourth Guided Project from the DataQuest Data Science track. The goal is to explore a data set consisting of college majors and their respective statistics in the job market.  

This guided project introduces the use Data Visualiztion methods with Matplotlib. 

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1. Review the Data Set

```python
recent_grads = pd.read_csv('recent-grads.csv')
recent_grads = pd.DataFrame(recent_grads)
print(recent_grads.iloc[0,:])
print(recent_grads.head())
print(recent_grads.tail())
print(recent_grads.describe())
print('Original row count ',len(recent_grads))
recent_grads = recent_grads.dropna()
print('Row count after dropna ',len(recent_grads))
```

## 2. Scatter Plots

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Median vs. Sample_size', figsize=(5,5))
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Unemployment Rate vs. Sample_size', figsize=(5,5))
```

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Median vs. Full time', figsize=(5,5))
```

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='Unemployment Rate vs. Share Women', figsize=(5,5))
```

```python
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Median vs. Men', figsize=(5,5))
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Median vs. Women', figsize=(5,5))
```

Do students in more popular majors make more money?
Do students that majored in subjects that were majority female make more money?
Is there any link between the number of full-time employees and median salary?
